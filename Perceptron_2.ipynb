```python
import matplotlib.pyplot as plt
from Perceptron_ import Perceptron
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X,y=datasets.make_blobs(n_samples=1000,n_features=2,centers=2,cluster_std=2,random_state=2)
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=200)
```

```python
p = Perceptron(learning_rate=0.01, n_iters=1000)
p.fit(Xtrain, ytrain)
ypredict = p.predict(Xtest)
```

```python
accuracy_score(ytest,ypredict)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.scatter(Xtrain[:, 0], Xtrain[:, 1], marker="o", c=ytrain)

x0_1 = np.amin(Xtrain[:, 0])
x0_2 = np.amax(Xtrain[:, 0])

x1_1 = (-p.weights[0] * x0_1 - p.bias) / p.weights[1]
x1_2 = (-p.weights[0] * x0_2 - p.bias) / p.weights[1]

ax.plot([x0_1, x0_2], [x1_1, x1_2], "k")

ymin = np.amin(Xtrain[:, 1])
ymax = np.amax(Xtrain[:, 1])
ax.set_ylim([ymin - 3, ymax + 3])
```
