```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import investpy
import cufflinks as cf
cf.go_offline()
import plotly.io as pio 
```

```python
MBB=investpy.get_stock_historical_data(stock='MBB',country='vietnam',from_date='01/01/2016',to_date=dt.datetime.now().strftime('%d/%m/%Y'))
MBB
```

```python
MBB=MBB.drop('Currency',axis=1)
MBB['20D']=MBB['Close'].rolling(20).mean()
MBB['50D']=MBB['Close'].rolling(50).mean()
MBB['upper']=MBB['20D']+2*(MBB['Close'].rolling(20).std())
MBB['lower']=MBB['20D']-2*(MBB['Close'].rolling(20).std())
MBB['simple return']=MBB['Close']/MBB['Close'].shift(1)-1
MBB
```

```python
MBB['simple return'].mean()*252
```

```python
MBB['simple return'].tail(252).sum()
```

```python
MBB['simple return'].tail(252).var()
```

```python
MBB.sort_values(by='Close',ascending=False)
```

```python
MBB[['Close','20D','upper','lower']].tail(252).iplot(title='Bollinger Band')
#biểu đồ bollinger band
```

```python
MBB[['Close','20D','50D']].tail(252).iplot(title='CLOSE, MA20, MA50')
#biểu đồ giá đóng cửa, MA20 và MA50
```
